```python
import pandas as pd
import datetime 
```

```python
df_base = datetime.datetime.today()
```

```python
# create a list of dates with 20 days different from today .
df_date_list = [df_base-datetime.timedelta(days=x) for x in range (0,20)]
df_date_list[:5]
```

```python
# Now lets make it a dataframe
df = pd.DataFrame(df_date_list,columns=['Day'])
```

```python
# Now we will add weeday name .
df['Day_of_week'] = df['Day'].dt.day_name()
df.sample(4)
```

```python
df.Day_of_week.unique()
```

```python
df['Ordinal_day'] = df.Day_of_week.map(week_days)
```

### Que : So ,how to handle this  ? ? ?

### Ans : We can make one binary variable for each of the 10 (or what number you want) most frequent labels only . This is equivalent to grouping all the others labls (with less frequent) under a new category & then tell them good_bye.

```python
#Step_1 : lets see frequency of values in 'X0' column
df.X0.value_counts().sort_values(ascending=False).head(10)
```

```python
#Step_2: MAke a list wuth most frequent top 10 categories.
top_10 = [x for x in df.X0.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
import numpy as np
```

```python
# Step_3 : Lets make 10 binary variables .
for i in top_10:
    df[i] = np.where(df['X0']==i,1,0)
df[['X0']+top_10]
```

```python
# step_4 : lets apply on all columns:
def one_hot_top_10(df,variable,top_10):
    for label in top_10:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
df =pd.read_csv(r"Datasets\mercedes.csv",usecols=['X0','X1','X2','X3','X4'])
one_hot_top_10(df,'X0',top_10)
df
```
